## Description of the Dataset
This dataset contain 5 columns and 150 rows and this dataset contain attributes such as price,bedrooms and bathrooms etc.I am going to analyze the data deeply and try to get some useful insights.

It includes 18 variables.These variables are described below:
1. date: The date when the data was recorded. This could indicate when the house was sold or listed.
2. price: The selling price of the house, which is the target variable to be predicted.
3. bedrooms: The number of bedrooms in the house.
4. bathrooms: The number of bathrooms in the house.
5. sqft_living: The living area in square feet.
6. sqft_lot: The total area of the lot in square feet.
7. floors: The number of floors in the house.
8. waterfront: A binary indicator of whether the house has a waterfront view (1 if yes, 0 if no).
9. view: A score indicating the quality of the view from the house.
10. condition: The condition of the house, usually represented as a score.
11. sqft_above: The square footage of the house above ground level.
12. sqft_basement: The square footage of the basement.
13. yr_built: The year the house was built.
14. yr_renovated: The year the house was last renovated.
15. street: The street address of the house.
16. city: The city where the house is located.
17. statezip: The state and ZIP code of the house.
18. country: The country where the house is located.

```python
df=df.drop("country",axis=1)
```

```python
 def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df


numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for column in numeric_cols:
    df = remove_outliers(df, column)
```

```python
df.info()
```

```python
df=df.drop("date",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
df["city"]=label.fit_transform(df["city"])
df["statezip"]=label.fit_transform(df["statezip"])
df["street"]=label.fit_transform(df["street"])
```

```python
x=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
y_pred=model1.predict(x_test)
```
