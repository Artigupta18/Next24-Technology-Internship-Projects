## Description of the Dataset
This dataset is focused on social media sentiment analysis and includes various attributes related to social media posts.

It includes 15 variables.These variables are described below:

1. Unnamed: 0.1: An integer column likely serving as an index or identifier for the rows. It could be a duplicate or a split version of the main index.
2. Unnamed: 0: Another integer column, possibly another index or identifier, similar to Unnamed: 0.1.
3. Text: The main content of the social media post, containing the textual data that will be analyzed for sentiment.
4. Sentiment: The sentiment label associated with the post, which could be 'Positive', 'Negative', or 'Neutral'. This is the target variable for classification.
5. Timestamp: The date and time when the post was made, useful for temporal analysis.
6. User: The username or identifier of the person who made the post.
7. Platform: The social media platform where the post was made, such as Twitter, Facebook, or Instagram.
8. Hashtags: The hashtags used in the post, which could provide additional context or keywords.
9. Retweets: The number of retweets or shares the post received, indicating its reach and engagement level.
10. Likes: The number of likes the post received, another measure of engagement.
11. Country: The country from where the post was made, useful for geographical analysis.
12. Year: The year when the post was made.
13. Month: The month when the post was made.
14. Day: The day when the post was made.
15. Hour: The hour when the post was made, useful for understanding posting patterns.

```python
positive_sentiments = [
    'Positive', 'Happiness', 'Joy', 'Love', 'Excitement', 'Admiration', 'Gratitude', 'Hope', 'Inspiration',
    'Contentment', 'Pride', 'Affection', 'Anticipation', 'Euphoria', 'Elated', 'Empathetic', 'Enthusiasm',
    'Satisfaction', 'Surprise', 'Wonder', 'Optimism', 'Kindness', 'Celebration', 'Achievement', 'Fulfillment',
    'Confidence', 'Harmony', 'Playful', 'Serenity', 'Compassion', 'Delight', 'Connection', 'Inspiration',
    'Triumph', 'Heartwarming'
]

negative_sentiments = [
    'Negative', 'Sadness', 'Anger', 'Fear', 'Frustration', 'Disgust', 'Despair', 'Grief', 'Loneliness',
    'Regret', 'Disappointment', 'Jealousy', 'Betrayal', 'Sorrow', 'Heartbreak', 'Anxiety', 'Confusion',
    'Helplessness', 'Overwhelmed', 'Resentment', 'Isolation', 'Bitterness', 'Numbness', 'Guilt', 'Shame',
    'Stress', 'Pain', 'Desperation', 'Boredom', 'Fearful', 'Melancholy', 'Desolation'
]

neutral_sentiments = [
    'Neutral', 'Calmness', 'Acceptance', 'Indifference', 'Serenity', 'Curiosity', 'Reflection', 'Ambivalence',
    'Mindfulness', 'Nostalgia', 'Introspection', 'Contemplation'
]
```

```python
def map_sentiment(sentiment):
    if sentiment in positive_sentiments:
        return 'Positive'
    elif sentiment in negative_sentiments:
        return 'Negative'
    elif sentiment in neutral_sentiments:
        return 'Neutral'
    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Sentiment'].apply(map_sentiment)
```

```python
df = df.drop(['Unnamed: 0.1','Unnamed: 0','Timestamp'],axis=1)
```

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'http\S+|www.\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#', '', text)  # Remove hashtags
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.strip()  # Remove whitespace
    stop_words = set(stopwords.words('english'))
    text = ' '.join(word for word in text.split() if word not in stop_words)
    
    lemmatizer = WordNetLemmatizer()
    text = ' '.join(lemmatizer.lemmatize(word) for word in text.split())
    
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['Text'] = df['Text'].apply(clean_text)

vectorizer = TfidfVectorizer(max_features=5000)
x = vectorizer.fit_transform(df['Text']).toarray()
```
